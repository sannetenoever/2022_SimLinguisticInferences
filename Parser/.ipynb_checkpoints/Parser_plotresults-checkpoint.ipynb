```python
## Setting up everything
import os
import sys
import matplotlib.pyplot as plt
import numpy as np
from numpy import matlib
uppath = os.path.dirname(os.getcwd())
sys.path.append(uppath)
sys.path.append(os.getcwd() + '/Code/')
import ColorScheme as cs
cmaps = cs.CCcolormap()
bfi = cs.baseFigInfo()   
import FFTanaSub as subfun
dataloc = os.getcwd() + '/Data/'
```

```python
## Function that does the FFT analysis + plotting
def doana(upsam = False, lay = 'all'):
    # shape of the data:
        # 20 - number of samples
        # 8 - number of layers
        # 52 - number of time steps per sample. Recall that we can only input one (valid) word at a time. So here we have 52/4 = 13 sequences in each sample. The sampling rate is 4 words per second.
        # 1024 - embedding dimension
        
    typesen = ['sen', 'wordlist', 'nounphrase', 'verbphrase','nongrammar', 'random']
    # wordlist = randomly chosen sequences of words. 
    # nongrammar is grammatical but meaningless (anvn). 
    # Random pattern is vnna
    
    # set up the plotting
    mifoi = 0.75 # minimum frequency to plot
    mfoi = 4.5 # maximum frequency to plot
    axs = list()
    fig = plt.figure(constrained_layout=True, figsize = (bfi.figsize.Col15,3))
    gs = fig.add_gridspec(2,3)
    
    # load the data
    s1hz = np.zeros([8,len(typesen)])
    for it, Stype in enumerate(typesen): 
        filename = Stype + '.npy'
        val = np.load(dataloc + filename)  
        
        # upsample if requested
        if upsam:
            val2 = np.zeros([20,8,52*2,1024])
            val2[:,:,np.arange(0,52*2,2),:] = val
            fs = 8
        else:
            val2=val
            fs = 4
        
        # figure out the axis of the plot
        if it < 3:
            axs.append(fig.add_subplot(gs[0,it])) 
        else:
            axs.append(fig.add_subplot(gs[1,it-3])) 
        
        # plot either all layers or only the last layer of the Parser
        if lay == 'all':
            pl = np.arange(0,val.shape[1])            
        elif lay == 'last':
            pl = [val.shape[1]-1]
        for L in pl:   
            # do the fft analysis
            x = np.squeeze(val2[:,L,:,:])
            act = np.moveaxis(x, [0,1,2], [0,1,2])            
            ft,freqs = subfun.ffthan_wrap(act, avgHidden = False, evoked = False, z = False, sr = fs)
            
            # get right frequencies
            frinx = (freqs > mifoi) & (freqs < mfoi)
            freqs = freqs[frinx]
            ft = ft[frinx]                                 
             
            # plot
            axs[-1].plot(freqs,ft) 
            
            # get the 1 Hz power
            fri = np.argmin(np.abs(freqs-1))
            s1hz[L,it] = ft[fri]
        
        # fix axis
        axs[-1].set_ylim([-0.5, 0.5])
        axs[-1].set_ylim([0, 10])
        axs[-1].set_xlim([0.75, 4.25])
        axs[-1].set_xticks([1,2,3,4])
        axs[-1].set_title(Stype)    
        if it == 0:
            axs[-1].legend(np.arange(0,8))
        if (it == 0) | (it == 3):
            axs[-1].set_ylabel('power (z-score)')
        else:
            axs[-1].get_yaxis().set_visible(False)
        if it > 2:
            axs[-1].set_xlabel('freq (Hz)')
        else:
            axs[-1].get_xaxis().set_visible(False)
    fig.tight_layout()    
    plt.show()
    
    # if all layers make separate plot of 1 Hz response
    if lay == 'all':
        fig = plt.figure(constrained_layout=True, figsize = (bfi.figsize.Col1,1.5))
        gs = fig.add_gridspec(1,2)
        axs = []
        axs.append(fig.add_subplot(gs[0,0])) 
        axs[-1].plot(s1hz)
        axs[-1].legend(typesen)
        axs[-1].set_ylabel('power (z-score)')
        axs[-1].set_xlabel('layer depth')
        plt.show()
```

```python
# Do the analysis and plot

# upsam = True or False (4 or 8 Hz)
# lay = all or last (all layers or last layer)

doana(upsam = False, lay = 'all')
```
